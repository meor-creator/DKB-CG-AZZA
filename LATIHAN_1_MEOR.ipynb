## LATIHAN 1 MEOR

### PROGRAM MEMINTA NAMA DAN UMUR

```python
nama = input("Masukkan nama anda:")
umur = int(input("Masukkan umur anda:"))

print("Nama:", nama)
print("Umur:", umur, "tahun")
```

### PROGRAM MENGIRA LUAS SEGI EMPAT

```python
panjang = float(input("Masukkan panjang: "))
lebar = float(input("Masukkan lebar: "))

luas = panjang * lebar

print("Luas segi empat ialah:", luas)
```

### PROGRAM MENENTUKAN NOMBOR GANJIL ATAU GENAP

```python
nombor = int(input("Masukkan satu nombor: "))

if nombor % 2 == 0:
    print("Nombor genap")
else:
    print("Nombor ganjil")
```

### PAPARKAN NOMBOR 1 HINGGA 10

```python
for i in range(1, 11):
    print(i)
```

### PROGRAM MENENTUKAN LULUS ATAU GAGAL

```python
markah = int(input("Masukkan markah: "))

if markah >= 50:
    print("Status: Lulus")
else:
    print("Status: Gagal")
```
